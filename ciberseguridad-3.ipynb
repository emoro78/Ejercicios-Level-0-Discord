```python
# Importar librerias
import getpass

# Almacenar las franjas horarias disponibles para cada programa en los distintos campus
    'London': {'Computer Science': 1, 'Medicine': 1, 'Marketing': 1, 'Arts': 1},
    'Manchester': {'Computer Science': 3, 'Medicine': 3, 'Marketing': 3, 'Arts': 3},
    'Liverpool': {'Computer Science': 1, 'Medicine': 1, 'Marketing': 1, 'Arts': 1}
}

# Almacenar los usuarios registrados
registered_users = {}

# Almacenar los intentos de inicio de sesión de cada usuario
login_attempts = {}

def login():
    
    # Solicitar al usuario usuario y contraseña
    username = input("Enter your username: ")
    password = getpass.getpass("Enter your password: ")

    # Comprobar si el usuario ha superado el máximo de intentos de inicio de sesión
    if username not in login_attempts:
        login_attempts[username] = 0

    if login_attempts[username] >= 3:
        print("Account locked. Please contact the administrator.")
        return False

    # Comprobar si el usuario y la contraseña son validos
    if username in registered_users and registered_users[username]['password'] == password:
        login_attempts[username] = 0
        return True
    else:
        login_attempts[username] += 1
        print("Invalid username or password")
        return False

def display_programs():
    
    # Mostrar al usuario los programas disponibles
    print("\nAvailable Programs:")
    print("1. Computer Science")
    print("2. Medicine")
    print("3. Marketing")
    print("4. Arts")

def enroll_program(username):
    
    # Pedir al usuario que introduzca el programa en el que desea inscribirse
    program = input("Enter the program you want to enroll in: ")

    # Comprobar si el programa es válido y está disponible en el campus elegido
    if program not in available_slots[city]:
        print("Invalid program")
        return

    if available_slots[city][program] > 0:
        
        # Inscribir al usuario en el programa y actualizar las franjas horarias disponibles
        registered_users[username]['program'] = program
        available_slots[city][program] -= 1
        print("Enrollment successful")
    else:
        
        # ofrecer al usuaroi la opción de inscribirse en otro campus
        print("Program is unavailable at this campus")
        enroll_another_campus(username, program)

def enroll_another_campus(username, program):
    
    # Preguntar al usuario si desea matricularse en el programa en otro campus
    print("Do you want to enroll in the program at another campus?")
    choice = input("Enter 'yes' to enroll at another campus, or 'no' to cancel: ")

    if choice.lower() == 'yes':
        
        # Comprobar si el programa está disponible en algún otro campus e inscribir al usuario
        for campus, slots in available_slots.items():
            if slots[program] > 0:
                registered_users[username]['program'] = program
                available_slots[campus][program] -= 1
                print("Enrollment successful at", campus)
                return

        print("Program is unavailable at all campuses")
    else:
        print("Enrollment canceled")

def enrollment_system():
    
    # Mostrar el menu del sistema
    print("University Enrollment System")
    print("-----------------------------")

    while True:
        if not login():
            continue

        display_programs()
        program_choice = input("Enter the program number you want to enroll in: ")

        # Verificar si la opción elegida por el usuario es valida
        if program_choice not in ['1', '2', '3', '4']:
            print("Invalid program choice")
            continue

        # Ingresar datos del usuario
        first_name = input("Enter your first name: ")
        last_name = input("Enter your last name: ")

        city = input("Choose a campus (London, Manchester, Liverpool): ")

        if city not in available_slots:
            print("Invalid campus choice")
            continue

        if available_slots[city]['Computer Science'] == 0 and available_slots[city]['Medicine'] == 0 and \
                available_slots[city]['Marketing'] == 0 and available_slots[city]['Arts'] == 0:
            print("No available slots at this campus")
            continue

        username = first_name.lower() + last_name.lower()
        password = getpass.getpass("Create a password: ")

        registered_users[username] = {
            'first_name': first_name,
            'last_name': last_name,
            'password': password,
            'program': None
        }

        enroll_program(username)

        choice = input("Do you want to perform another operation? (yes/no): ")

        if choice.lower() != 'yes':
            print("Exiting...")
            break

enrollment_system()
```
