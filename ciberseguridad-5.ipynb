```python
# Diccionario para almacenar los ingresos y gastos del usuario
finance_data = {
    'income': 0,
    'expenses': {
        'Medical expenses': 0,
        'Household expenses': 0,
        'Leisure': 0,
        'Savings': 0,
        'Education': 0
    }
}

def record_income():
    
    # Registrar el valor ingresado por el usuario
    income = float(input("Enter your total income: "))
    finance_data['income'] = income

def record_expense():
    
    # Registrar los gastos del usuario
    print("\nExpense Categories:")
    print("1. Medical expenses")
    print("2. Household expenses")
    print("3. Leisure")
    print("4. Savings")
    print("5. Education")

    category_choice = input("Enter the category number for the expense: ")

    if category_choice not in ['1', '2', '3', '4', '5']:
        print("Invalid category choice")
        return

    expense_amount = float(input("Enter the expense amount: "))

    category = None
    if category_choice == '1':
        category = 'Medical expenses'
    elif category_choice == '2':
        category = 'Household expenses'
    elif category_choice == '3':
        category = 'Leisure'
    elif category_choice == '4':
        category = 'Savings'
    elif category_choice == '5':
        category = 'Education'

    finance_data['expenses'][category] += expense_amount

def calculate_total_expenses():
    
    # Calcular los gastos totales
    total_expenses = sum(finance_data['expenses'].values())
    return total_expenses

def calculate_category_total(category):
    
    # Calcular los gastos totales de una categoría específica
    category_total = finance_data['expenses'][category]
    return category_total


def finance_management():
    
    # Main para gestionar la aplicación
    print("Finance Management Application")
    print("------------------------------")

    record_income()

    while True:
        
        # Mostrar el menu al usuario
        print("\nMenu:")
        print("1. Record Expense")
        print("2. Get Total Expenses")
        print("3. Get Category Total")
        print("4. Exit")

        # Solicitar el ingreso de la opcion a elegir
        choice = input("Enter your choice: ")

        # Verificar que la opcion sea valida
        if choice == "1":
            record_expense()
        elif choice == "2":
            total_expenses = calculate_total_expenses()
            print("Total Expenses: $", total_expenses)

            if total_expenses == finance_data['income']:
                print("You have spent all your income. Consider reducing expenses in the category where you have spent the most money.")
            elif total_expenses < finance_data['income']:
                print("Congratulations! You have spent less than your income.")
            else:
                print("You have spent more than your income. Consider improving your financial health.")
        elif choice == "3":
            print("\nExpense Categories:")
            print("1. Medical expenses")
            print("2. Household expenses")
            print("3. Leisure")
            print("4. Savings")
            print("5. Education")

            category_choice = input("Enter the category number: ")

            if category_choice not in ['1', '2', '3', '4', '5']:
                print("Invalid category choice")
                continue

            category = None
            if category_choice == '1':
                category = 'Medical expenses'
            elif category_choice == '2':
                category = 'Household expenses'
            elif category_choice == '3':
                category = 'Leisure'
            elif category_choice == '4':
                category = 'Savings'
            elif category_choice == '5':
                category = 'Education'

            category_total = calculate_category_total(category)
            print("Total for", category, ": $", category_total)
        elif choice == "4":
            print("Exiting...")
            break
        else:
            print("Invalid choice")

finance_management()
```
