```python
# Diccionario de tipos de cambio de distintas monedas
exchange_rates = {
    'CLP': 0.0012,
    'ARS': 0.010,
    'USD': 1.0,
    'EUR': 1.18,
    'TRY': 0.12,
    'GBP': 1.38
}

# Función de conversión de divisas
def currency_converter():
    print("Currency Converter")
    print("------------------")

    while True:
        
        # Mostrar el menu al usuario
        print("\nMain Menu:")
        print("1. Convert Currency")
        print("2. Exit")

        choice = input("Enter your choice: ")

        # Verificar si la opcion dada por el usuario es valida
        if choice == "1":
            initial_currency = input("Enter your initial currency: ")
            target_currency = input("Enter the currency you want to exchange to: ")

            # Comprobar si las divisas introducidas son válidas
            if initial_currency not in exchange_rates or target_currency not in exchange_rates:
                print("Invalid currency")
                continue

            amount = float(input("Enter the amount to convert: "))

            # Comprobar si el valor introducido es válido
            if amount < 0:
                print("Invalid amount")
                continue

            min_amount = 10  # Valor mínimo para cada divisa
            max_amount = 10000  # Valor maximo para cada divisa

            # Compruebe si el valor introducido se encuentra dentro del intervalo válido
            if amount < min_amount or amount > max_amount:
                print("Amount must be between", min_amount, "and", max_amount)
                continue

            exchange_rate = exchange_rates[target_currency]
            converted_amount = amount * exchange_rate

            print("Converted amount:", converted_amount)

            withdraw = input("Do you want to withdraw the funds? (yes/no): ")

            if withdraw.lower() == "yes":
                commission = converted_amount * 0.01
                converted_amount -= commission
                print("Withdrawn amount:", converted_amount)
                print("Commission:", commission)

        elif choice == "2":
            print("Exiting...")
            break

        else:
            print("Invalid choice")

# Llama a la función currency_converter para iniciar el programa
currency_converter()
```
